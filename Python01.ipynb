# **ตัวอย่าง โค้ดการตัดเกรด จากตัวแปร score**

```python
# ทบทวนตัวแปร grade ตัวเต็ม
grade
```

```python
score = 75
if score >= grade["A"]:
  print("A")
elif score >= grade["B+"]:
  print("B+")
elif score >= grade["B"]:
  print("B")
elif score >= grade["C"]:
  print("C")
elif score >= grade["D"]:
  print("D")
else:
  print("F")
```
