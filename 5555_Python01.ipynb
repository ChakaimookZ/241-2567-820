```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#อ่านข้อมูลจากไฟล์ study.csv
df = pd.read_csv('study.csv')
df.head()
```

```python
#ตรวจสอบว่ามีคอลัมอะไรบ้าง
df.columns
```

```python
input_columns = ['Hours'] #แก้
output_columns = ['Scores'] #แก้
```

```python
X = df[input_columns]
y = df[output_columns]
```

```python
#X.shape
```

```python
# splitting the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.20, random_state=1,)
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
model = LinearRegression() #สร้าง model : y=mx + b
model.fit(X_train, y_train) # เรียน หาค่า m กับ b ออกมา
```

```python
# แสดงค่า m และ b
print(model.intercept_) # b จุดตัดแกน y
print(model.coef_) # m ความชัน
```

```python
#ทำนายบนชุดทดสอบ
y_p = model.predict(X_test)
```

```python
#วัดผล MSE
from sklearn import metrics
print(np.sqrt(metrics.mean_squared_error(y_test, y_p)))
```

```python
#หากเรียน 7.5,8.5,9.25
X_quiz = [[7.5], [8.5], [9.25]]
y_quiz = model.predict(X_quiz)
print(y_quiz)
```

```python
#ใช้ model  ที่ได้ หากเรียน 12.2,15.3,1,0,6,7.2
#จะได้คะแนนเท่าใด
X_quiz = [[12.2], [15.3], [1], [0], [6], [7.2]]
y_quiz = model.predict(X_quiz)
print(y_quiz)
```

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
#อ่านข้อมูลจากไฟล์ study.csv
df = pd.read_csv('insurance.csv')
df.head()
```

```python
#ตรวจสอบว่ามีคอลัมอะไรบ้าง
df.columns
```

```python
#กำจัดค่าว่าง
df.dropna(inplace = True )
```

```python
input_columns = ['age', 'bmi', 'children'] #แก้
output_columns = ['charges'] #แก้
```

```python
X = df.loc[:, input_columns]
y = df.loc[:, output_columns]
```

```python
# splitting the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.20, random_state=1,)
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
model = LinearRegression() #สร้าง model : y=mx + b
model.fit(X_train, y_train) # เรียน หาค่า m กับ b ออกมา
```

```python
# แสดงค่า m และ b
print(model.intercept_) # b จุดตัดแกน y
print(model.coef_) # m ความชัน
```

```python
#ทำนายบนชุดทดสอบ
y_p = model.predict(X_test)
```

```python
#วัดผล MSE
from sklearn import metrics
print(np.sqrt(metrics.mean_squared_error(y_test, y_p)))
```

```python
#ใช้ model ที่ได้ ทำนายเบี้ยประกัน #แก้
#age = 30, bmi=35, childen = 1
#age = 19, bmi=25, childen = 2
#age = 41, bmi=23, childen = 0
X_quiz = [[30, 35, 1], [19, 25, 2], [41, 23, 0]]
y_quiz = model.predict(X_quiz)
print(y_quiz)
```
